```python
#Merge, Concat : Similar to SQL join query
df = pd.DataFrame(np.random.randn(10, 4))
pieces = [df.iloc[:3,[0,1]], df.iloc[3:7,[3,1]], df.iloc[7:,[0,2]]]
new_df = pd.concat(pieces)
new_df
```

```python
students = {
    "id":[1,2,3,5],
    "Name":["Vedant","A","B","C"]
}
marks = {
    "id" : range(1,5),
    "CGPA":np.random.rand(4)*10
}
stu = pd.DataFrame(students)
stu= stu.set_index('id')
mark = pd.DataFrame(marks)
mark = mark.set_index('id')
new_df = pd.merge(stu,mark,on="id",how="outer")
# new_df.shift(1)
new_df
```

```python
# Grouping
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
print(df)
print(df.groupby(['B','A']).max()['C'])
```
